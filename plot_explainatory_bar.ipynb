```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
alphaerr = 0.2
# All variances
cor = np.random.randn(9000000)*0.45
gg = np.random.randn(9000000)*0.4
wd = gg = np.random.randn(9000000)*0.2
#
a= np.random.randn(9000000)*0.35
sae = np.random.randn(9000000)*0.3
sa = -a +sae #full correlated part + error from the sum of other
sac=a+sa

#
sb = np.random.randn(9000000)*sa.std()
sbc= a+sb

# intermediaire 
cerr = np.random.randn(9000000)*0.1*a.std()
c = a*a.std() + cerr
sc = -c + sae


```

```python
# before adding term 

fig, ax = plt.subplots(nrows=1,ncols=1,figsize=(7,4), sharey=True)

ax.bar(0.5, wd.var(), color ='mediumvioletred',width = 0.4, zorder=3)
ax.bar(0.5, gg.var(), color ='c',width = 0.4, zorder=3, bottom= wd.var())
ax.bar(0.5, cor.var(), color ='g',width = 0.4, zorder=3, bottom= wd.var()+gg.var())

ax.bar(1, sa.var(), color ='grey',width = 0.4, zorder=3)

ax.grid()
ax.set_xlim(0.2,2.7)
ax.set_xlim(0.27,2.26)
ax.set_ylim(0, 0.42)

ticks = (r'$\sum_{j\neq i} \langle x_j^2\rangle $',r'$\langle S_{-x_i}^2\rangle$')
ax.set_xticks([0.5, 1], ticks,)
#ax.set_yticks()
ax.set_yticklabels([])

ax.set_ylabel(r'$\langle ....^2 \rangle$ $[m^2/s^4]$')
```

```python
# before adding term 

fig, ax = plt.subplots(nrows=1,ncols=1,figsize=(7,4), sharey=True)

ax.bar(0.5, wd.var(), color ='mediumvioletred',width = 0.4, zorder=3)
ax.bar(0.5, gg.var(), color ='c',width = 0.4, zorder=3, bottom= wd.var())
ax.bar(0.5, cor.var(), color ='g',width = 0.4, zorder=3, bottom= wd.var()+gg.var())
ax.bar(0.5, a.var(), color ='r',width = 0.4, zorder=3, bottom = wd.var()+gg.var()+cor.var())

ax.bar(1, sa.var(), color ='grey',width = 0.4, zorder=3)
ax.bar(1, a.var(), color ='r',width = 0.4, zorder=3, bottom = sa.var())
ax.annotate(r'$\langle x^2 \rangle$', (1,sa.var()+0.5*a.var()),ha='center')
ax.annotate(r'$\langle S_{-x_i}^2 \rangle$', (1,sa.var()*0.5),ha='center')

ax.bar(1.5, sac.var(), color ='k',width = 0.4, zorder=3)
ax.axhline(sac.var(), color ='k', ls='--', zorder=4)
ax.annotate('Fully explained \n physical signal \n' +r'$\langle x^2 \rangle$ = $\langle \hat{x}^2 \rangle$', (1.5,sac.var()+0.01),ha='center')

ax.annotate(r'$RC_x = min $', (0.5, sac.var()+0.01),ha='center')

#ax.bar(2, sbc.var(), color ='k',width = 0.4, zorder=3)
#ax.axhline(sbc.var(), color ='k', ls='--', zorder=4)

ax.grid()
ax.set_xlim(0.27,2.26)
ax.set_ylim(0, 0.42)

ticks = (r'$\sum_{j\neq i} \langle x_j^2\rangle + \langle x_i^2\rangle$',r'$\langle S_{-x_i}^2\rangle+ \langle x_i^2\rangle$', r'$\langle S^2 \rangle$' )
ax.set_xticks([0.5, 1, 1.5], ticks,)

ax.set_yticklabels([])

ax.set_ylabel(r'$\langle ....^2 \rangle$ $[m^2/s^4]$')
```
