# Read Datacube

Notebook pour lire les datacubes à partir du catalogue intake (Utiliser l'environnement *data-env*)

```python
import xarray as xr
import intake

from pathlib import Path
```

```python
dataset = 'aviso' # Nom du jeu de données (aviso / erastar)
name = 'datacube-1' # Nom du datacube
```

```python
WRK = Path(f'/home1/datawork/gcaer/data/{dataset}/{name}')
```

```python
# Définir la taille des chunks
chunks = dict(
    time=1,
    lat=-1,
    lon=-1
)
```

```python
# Liste des variables à supprimer
_drop = [
    "crs",
    "lat_bnds",
    "lon_bnds",
    "ugosa",
    "err_ugosa",
    "vgosa",
    "err_vgosa",
    "ugos",
    "vgos",
    "flag_ice",
    "tpa_correction",
    "nv",
]
```

```python
%%time
cat = intake.open_catalog(WRK / f"references/aviso/reference.yaml")
ds = cat["data"](chunks={"time": chunks['time'], "latitude": chunks['lat'], "longitude": chunks['lon']}).to_dask()
ds = ds.drop_vars(_drop)
```
