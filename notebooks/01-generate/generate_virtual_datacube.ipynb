# Generate a virtual datacube 

(Utiliser l'environnement *data-env*)

Ce notebook présente comment générer un datacube virtuel avec des valeurs aléatoires. (*L'intérêt ici est d'avoir la donnée en mémoire et ne pas aller lire des fichiers.*)

```python
import numpy as np
import pandas as pd
import xarray as xr
```

## Datacube

```python
%%time
latitudes = np.linspace(-90, 90, 720)
longitudes = np.linspace(-180, 180, 1440)
times = pd.date_range('2000-01-01', periods=365, freq='D')
data = np.random.randn(len(times), len(latitudes), len(longitudes))
```

```python
datacube = xr.Dataset(
    data_vars=dict(data=(['time', 'latitude', 'longitude'], data)),
    coords=dict(
        latitude=latitudes,
        longitude=longitudes,
        time=times,
    )
)
```
