## Load data  and add local coordinates oriented along the altimetric track and along track sla gradient

Need to chunk a bit along `obs` dimension

- `site_obs`: drifter time
- `sassa_time`: satellite time


```python
labels=cstes.labels
```

```python
DS= {l: xr.open_zarr(cstes.zarr_dir+'/'+l+'.zarr').persist()for l in labels}
```

```python
dl = 2
lon, lat = np.arange(-180, 180, dl), np.arange(-90,90,dl)
acc_bins=np.arange(-1e-4,1e-4,1e-6)

diag_list = [ diag.diag_ind_pdfs, diag.diag_combined_pdfs,diag.diag_coloc_repartition]
bins_list = [acc_bins]*2+[[lon,lat]]
```

```python
DS_diag = diag.DS_Diag(DS, diag_list, bins_list)
```

## PDF

### Acceleration

```python
fig = plt.figure(figsize=(15,8))
j=0
n=len(DS_diag['drifter_sat_year'])
for i in range (n):
    ax = fig.add_subplot(2,int(np.ceil(n/2)),j+1)
    DS_diag.isel(drifter_sat_year = i).pdf_acc_x.plot( color='red', label=r"$d_tv_x$", ax=ax)
    DS_diag.isel(drifter_sat_year = i).pdf_acc_y.plot( color='teal', label=r"$d_tv_y$", ax=ax)
    ax.set_title(DS_diag.drifter_sat_year.values[i])
    j+=1
    ax.grid()
ax.legend(loc='center right',bbox_to_anchor=(1.5, 0.8))

fig.suptitle(r"Pdf of the acceleration term $d_t\vec{u}$ pdf in the local frame (x,y)")
fig.tight_layout()
```

## Coriolis

```python
fig = plt.figure(figsize=(15,8))
j=0
n=len(DS_diag['drifter_sat_year'])
for i in range (n):
    ax = fig.add_subplot(2,int(np.ceil(n/2)),j+1)
    DS_diag.isel(drifter_sat_year = i).pdf_coriolis_x.plot(color='red', label=r"-$fv_y$", ax=ax)
    DS_diag.isel(drifter_sat_year = i).pdf_coriolis_y.plot(color='teal',  label=r"$fv_x$",  ax=ax)
    ax.set_title(DS_diag.drifter_sat_year.values[i])
    j+=1
    ax.grid()
ax.legend(loc='center right',bbox_to_anchor=(1.5, 0.8))

fig.suptitle(r"Pdf of Coriolis acceleration term $f\vec{k}\times\vec{u}$ pdf in the local frame (x,y)")
fig.tight_layout()
```

### Pressure gradient 

```python
fig = plt.figure(figsize=(15,8))
j=0
n=len(DS_diag['drifter_sat_year'])
for i in range (n):
    ax = fig.add_subplot(2,int(np.ceil(n/2)),j+1)
    DS_diag.isel(drifter_sat_year = i).pdf_g_grad_sla.plot(color='orange',  label=r"$-g\vec{\nabla}(\eta)$ ", ax=ax)
    DS_diag.isel(drifter_sat_year = i).pdf_g_grad_sla_denoised.plot(color='grey',  label=r"$-g\vec{\nabla}(\eta)$ denoised",ax=ax)
    ax.set_title(DS_diag.drifter_sat_year.values[i])
    j+=1
    ax.grid()
ax.legend(loc='center right',bbox_to_anchor=(1.5, 0.8))

fig.suptitle(r"Pdf of the pressure gradient term $-g\vec{\nabla}(\eta$)")
fig.tight_layout()
```
