```python
zarr = os.path.join(zarr_dir, "stat_matchup_err_argos.zarr")
ds_argos=xr.open_zarr(zarr).persist()
zarr = os.path.join(zarr_dir, "stat_matchup_err_gps.zarr")
ds_gps=xr.open_zarr(zarr).persist()
```

# PLOT

```python
ds_gps
```

```python
fig, ax = plt.subplots(nrows=1,ncols=4, figsize=(12,3))
ax=ax.flatten()
ds_gps0= ds_gps.where(ds_gps.drifter_err_x<200).where(ds_gps.drifter_err_y<200)
ds_argos0= ds_argos.where(ds_argos.drifter_err_x<2e3).where(ds_argos.drifter_err_y<2e3)
b=200
mean = np.ceil([ds_gps0.drifter_err_x.mean().values,
                ds_gps0.drifter_err_y.mean().values,
                ds_argos0.drifter_err_x.mean().values,
                ds_argos0.drifter_err_y.mean().values,])

ds_gps0.drifter_err_x.plot.hist(bins=b, ax=ax[0],density=True, color='salmon')
ds_gps0.drifter_err_y.plot.hist(bins=b, ax=ax[1],density=True, color='salmon')
ds_argos0.drifter_err_x.plot.hist(bins=b, ax=ax[2],density=True, color='firebrick')
ds_argos0.drifter_err_y.plot.hist(bins=b, ax=ax[3],density=True, color='firebrick')
#fig.legend(ncol=4, fontsize=10,loc='lower center')#, bbox_to_anchor=(1,         0.5))#bbox_transform=fig.transFigure,
for i in range(len(ax)) :
    #ax[i].set_yscale('log')
    ax[i].set_axisbelow(True)
    ax[i].set_title(cstes.lettres[i]+f'\n mean ={mean[i]}')
    ax[i].grid()
ax[0].set_xlim(0, 200)
ax[1].set_xlim(0, 200)
ax[2].set_xlim(0, 2e3)
ax[3].set_xlim(0, 2e3)
fig.tight_layout(rect=[0,0.1,1,1])#left, bottom, right, top (default is 0,0,1,1)
```

```python
err_vx_gps = (ds_gps0.drifter_err_x*np.sqrt(2)/3600).assign_attrs({"long_name":r'$err_{vx}$', "units":'m/s'})
err_vy_gps = (ds_gps0.drifter_err_y*np.sqrt(2)/3600).assign_attrs({"long_name":r'$err_{vx}$', "units":'m/s'})
err_vx_argos = (ds_argos0.drifter_err_x*np.sqrt(2)/3600).assign_attrs({"long_name":r'$err_{vx}$', "units":'m/s'})
err_vy_argos = (ds_argos0.drifter_err_y*np.sqrt(2)/3600).assign_attrs({"long_name":r'$err_{vx}$', "units":'m/s'})
```
