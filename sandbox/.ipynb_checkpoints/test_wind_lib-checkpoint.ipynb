## Compute relevant aviso related variables

```python
def get_era5_one_obs(ds_obs, dt):
    """ load aviso for one collocation """
    
    # load data
    era5 = wind.load_era5(ds_obs.time.values, dt, suffix="era5_")# aviso for matchup time
    
    # interpolate on local grid
    u10 = era5.u10.interp(era5_lon=ds_obs.box_lon,
                           era5_lat=ds_obs.box_lat,
                          )
    v10 = era5.v10.interp(era5_lon=ds_obs.box_lon,
                           era5_lat=ds_obs.box_lat,
                          )
    """
    err_sla = aviso.err_sla.interp(aviso_lon=ds_obs.box_lon,
                           aviso_lat=ds_obs.box_lat,
                          )
    """
    
    # convert to dataset and massage in order to concatenate properly in get_aviso
    #ds = sla.to_dataset()
    ds = xr.Dataset()
    ds['era5_u10'] = u10
    ds = ds.drop("time")
    ds["era5_dates"] = ds.era5_time
    ds['time']=xr.DataArray(ds_obs.time.values)
    
    
   
    ds = (ds
         .drop(["era5_time", "lon", "lat", "era5_lon", "era5_lat"])#"time", # forget about actual dates in time coordinate
   )#.rename(sla="aviso_sla")
    
    ds = ds.assign_coords({"era5_time":[_dt for _dt in range(*dt)] })
    ds.aviso_time.attrs = {'description' : 'delta time from the matchup time, 0 is the matchup time indice on aviso_time dimension', 'units':'days'}
    
    return ds

def get_aviso(ds, dt=None):
    """ Load aviso data for multiple collocations and concatenate """
    return xr.concat([get_aviso_one_obs(ds.sel(obs=o), dt) for o in ds.obs], 'obs')

def compute_aviso_sla(ds, dt=None):
    """
    https://xarray.pydata.org/en/stable/user-guide/dask.html
    Parameters
    ----------
    ds: xr.Dataset
        Input collocation dataset
    dt: tuple
        Time offsets compared to the collocation, e.g. if collocation
        is at time t and dt=(-5,5), aviso data will be interpolated on 
        the interval (t-5 days, t+5 days)
        Default is (-5,5)

    Return
    ------
    ds: xr.Dataset
        Dataset containing relevant aviso data

    """
    if dt is None:
        dt = (-5,5)

    # build template
    template = get_aviso(ds.isel(obs=slice(0,2)), dt=dt)

    # broad cast along obs dimension
    template, _ = xr.broadcast(template.isel(obs=0).chunk(), 
                               ds,
                               exclude=["aviso_time", "alti_time", "alti_time_mid", "site_obs", "box_x", "box_y",],
                          )
    
     # massage further and unify chunks
    template = (xr.merge([template, ds.drop(["lon", "lat", "time"])])
            .unify_chunks()
            [["aviso_sla", "aviso_g_grad_x", "aviso_g_grad_y",
              "aviso_drifter_matchup_sla", "aviso_drifter_matchup_g_grad_x","aviso_drifter_matchup_g_grad_y",
              "aviso_alti_matchup_sla", "aviso_alti_matchup_g_grad_x","aviso_alti_matchup_g_grad_y","time","aviso_dates",
             ]]
            )
    
    # dimension order is not consistent with that of exiting from map_blocks for some reason
    template["aviso_sla"] = template["aviso_sla"].transpose("obs","aviso_time", "box_y", "box_x")
    template["aviso_g_grad_x"] = template["aviso_g_grad_x"].transpose("obs","aviso_time", "box_y", "box_x")
    template["aviso_g_grad_y"] = template["aviso_g_grad_y"].transpose("obs","aviso_time", "box_y", "box_x")
                                              
    #chunk alti_time et site_obs in only one chunk both
    ds=ds.chunk({'alti_time':ds_data.alti_time.size, 'site_obs':ds_data.site_obs.size})
    
    # actually perform the calculation
    ds_sla = xr.map_blocks(get_aviso, ds.unify_chunks(), kwargs=dict(dt=dt), template=template)
    ds_sla = ds_sla.set_coords("time")
    return ds_sla
```
