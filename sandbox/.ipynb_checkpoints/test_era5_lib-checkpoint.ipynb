## Compute relevant aviso related variables

```python
def get_era5_one_obs(ds_obs, dt):
    """ load era5 for one collocation """
    
    # load data
    era5 = wind.load_era5(ds_obs.time.values, dt, suffix="era5_")# aviso for matchup time
    
    # interpolate on local grid
    u10 = era5.u10.interp(era5_lon=ds_obs.box_lon,
                           era5_lat=ds_obs.box_lat,
                          )
    v10 = era5.v10.interp(era5_lon=ds_obs.box_lon,
                           era5_lat=ds_obs.box_lat,
                          )
    """
    err_sla = aviso.err_sla.interp(aviso_lon=ds_obs.box_lon,
                           aviso_lat=ds_obs.box_lat,
                          )
    """
    
    # convert to dataset and massage in order to concatenate properly in get_era5
    #ds = sla.to_dataset()
    ds = xr.Dataset()
    ds['era5_u10'] = u10
    ds['era5_v10'] = v10
    ds = ds.drop("time")
    ds["era5_dates"] = ds.era5_time
    ds['time']=xr.DataArray(ds_obs.time.values)
    
    # matchup site drifter/sat

    site_matchup_indice = ds_obs.__site_matchup_indice.values
    drifter_x = ds_obs.drifter_x.isel(site_obs = site_matchup_indice).values
    drifter_y = ds_obs.drifter_y.isel(site_obs = site_matchup_indice).values
    drifter_time = ds_obs.drifter_time.isel(site_obs = site_matchup_indice).values
    alti_x = ds_obs.alti_x.isel(alti_time=10).values
    alti_y = ds_obs.alti_y.isel(alti_time=10).values
    alti_time = ds_obs.alti_time_.isel(alti_time=10).values
    
     # matchup site drifter/sat
    ds_matchup = ds_obs.isel(site_obs=ds_obs.__site_matchup_indice.compute(),
                             alti_time=10, alti_time_mid=10)
    
    
    ds['era5_drifter_matchup_u10'] = ds.era5_u10.interp(box_x = ds_matchup.drifter_x.values,
                                                           box_y = ds_matchup.drifter_y.values,
                                                          era5_time = ds_matchup.drifter_time.values
                                                         )
    ds['era5_drifter_matchup_v10'] = ds.era5_v10.interp(box_x = ds_matchup.drifter_x.values,
                                                                     box_y = ds_matchup.drifter_y.values,
                                                                     era5_time = ds_matchup.drifter_time.values
                                                                    )
    ds['era5_alti_matchup_u10'] = ds.era5_u10.interp(box_x = ds_matchup.alti_x.values,
                                                        box_y =ds_matchup.alti_y.values,
                                                       era5_time = ds_matchup.alti_time_.values
                                                       )
    ds['era5_alti_matchup_v10'] = ds.era5_v10.interp(box_x = ds_matchup.alti_x.values,
                                                                  box_y =ds_matchup.alti_y.values,
                                                                 era5_time = ds_matchup.alti_time_.values
                                                                 )
    
    ds = (ds
         .drop(["era5_time", "lon", "lat", "era5_lon", "era5_lat"])#"time", # forget about actual dates in time coordinate
   )#.rename(sla="aviso_sla")
    
    #ds = ds.assign_coords({"era5_time":np.concatenate([np.arange(_dt*24, (_dt+1)*24,1) for _dt in range(*dt)]) })
    ds.era5_time.attrs = {'description' : 'delta time from the matchup time, 0 is the matchup time indice on aviso_time dimension', 'units':'hours'}
    
    return ds

def get_era5(ds, dt=None):
    """ Load era5 data for multiple collocations and concatenate """
    return xr.concat([get_era5_one_obs(ds.sel(obs=o), dt) for o in ds.obs], 'obs')

def compute_era5(ds, dt=None):
    """
    https://xarray.pydata.org/en/stable/user-guide/dask.html
    Parameters
    ----------
    ds: xr.Dataset
        Input collocation dataset
    dt: tuple
        Time offsets compared to the collocation, e.g. if collocation
        is at time t and dt=(-5,5), aviso data will be interpolated on 
        the interval (t-5 days, t+5 days)
        Default is (-5,5)

    Return
    ------
    ds: xr.Dataset
        Dataset containing relevant era5 data

    """
    if dt is None:
        dt = (-5,5)

    # build template
    template = get_era5(ds.isel(obs=slice(0,2)), dt=dt)

    # broad cast along obs dimension
    template, _ = xr.broadcast(template.isel(obs=0).chunk(), 
                               ds,
                               exclude=["era5_time", "alti_time", "alti_time_mid", "site_obs", "box_x", "box_y",],
                          )
    
     # massage further and unify chunks
    template = (xr.merge([template, ds.drop(["lon", "lat", "time"])])#
            .unify_chunks()
            [["era5_u10", "era5_v10","time",'era5_drifter_matchup_u10','era5_drifter_matchup_v10','era5_alti_matchup_u10','era5_alti_matchup_v10',
              "era5_dates",
             ]]
            )
    
    # dimension order is not consistent with that of exiting from map_blocks for some reason
    template["era5_u10"] = template["era5_u10"].transpose("obs","era5_time", "box_y", "box_x")
    template["era5_v10"] = template["era5_v10"].transpose("obs","era5_time", "box_y", "box_x")
    
                                              
    #chunk alti_time et site_obs in only one chunk both
    ds=ds.chunk({'alti_time':ds_data.alti_time.size, 'site_obs':ds_data.site_obs.size})
    
    # actually perform the calculation
    ds_sla = xr.map_blocks(get_era5, ds.unify_chunks(), kwargs=dict(dt=dt), template=template)
    ds_sla = ds_sla.set_coords("time")
    return ds_sla
```
