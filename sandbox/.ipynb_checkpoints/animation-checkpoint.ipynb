```python
def plot_coloc_matchup(label, nobs, figsize =(10,6)):
    #dataset
    ds_data = xr.open_zarr(zarr_dir+'/'+label+'.zarr')
    ds_aviso =  xr.open_zarr(zarr_dir+'/aviso_'+label+'.zarr')
    ds_eras = xr.open_zarr(zarr_dir+'/erastar/erastar_'+label+'.zarr')
    
    ds_obs = ds_data.isel(obs=nobs).persist()
    #av_obs = aviso.get_aviso_one_obs(ds_obs, dt=(-30,31), only_matchup_time = False).set_coords('aviso_time_').set_index({'aviso_time':'aviso_time_'}).persist()
    av_obs = ds_aviso.isel(obs=nobs)['aviso_box_sla'].persist()
    era_obs = ds_eras.isel(obs=nobs)[['es_alti_matchup_taue', 'es_alti_matchup_taun',
                                      'es_drifter_matchup_taue', 'es_drifter_matchup_taun',]].persist()
    _ds =xr.merge([ds_obs, av_obs, era_obs])
    _dsm = _ds.isel(site_obs=int(_ds.__site_matchup_indice.values)).isel(alti_time_mid =10, alti_time =10).compute()

    lon, lat, t_matchup = float(_dsm.lon), float(_dsm.lat), _dsm.time.values
    td = pd.to_datetime(t_matchup)
    dt = float(_ds["alti___time_difference"])
    dl = float(_ds["alti___distance"])
    
    #figure
    fig = plt.figure(figsize=figsize)
    ax = fig.add_subplot(111, projection=ccrs.Orthographic(lon, lat))
    land_feature = cfeature.LAND

    # aviso
    min_=abs(_dsm["aviso_box_sla"].min())
    max_=abs(_dsm["aviso_box_sla"].max())
    a=max(min_,max_)
    C=ax.contourf(_dsm.box_lon, _dsm.box_lat, _dsm.aviso_box_sla, levels=30, cmap='bwr',transform=crs, vmin=-a, vmax=a)
    

    #axes
    lonc, latc=_dsm.box_lonc,_dsm.box_latc
    ax_y=_dsm.sel(box_x=0, box_y=50000)[['box_lon', 'box_lat']].compute()
    ax_x=_dsm.sel(box_x=100000, box_y=0)[['box_lon', 'box_lat']].compute()
    x_lon, x_lat= ax_x.box_lon,ax_x.box_lat
    y_lon, y_lat= ax_y.box_lon,ax_y.box_lat
    ax.arrow(lonc,latc, y_lon-lonc, y_lat-latc, color='black', lw=3,transform=crs,head_width=0.05)
    ax.arrow(lonc,latc, x_lon-lonc, x_lat-latc, color='black', lw=3,transform=crs,head_width=0.05)
    ax.text(y_lon, latc,'Y',transform=crs, fontsize=15)
    ax.text(lonc, x_lat,'X',transform=crs, fontsize=15)

    # drifter
    ax.scatter(_ds["drifter_lon"], _ds["drifter_lat"], c=np.arange(1, len(_ds["drifter_lat"].values)+1),cmap='viridis_r', marker='.',s=3, transform=crs, label="drifter's trajectory")
    ax.plot(_ds["drifter_lon"][0], _ds["drifter_lat"][0], "o", ms=5, color="y", transform=crs, label='start')
    ax.plot(_ds["drifter_lon"][-1], _ds["drifter_lat"][-1], "o", ms=10, color="navy", transform=crs, label='end')

    ax.plot(_dsm["drifter_lon"], _dsm["drifter_lat"], 
            "*", ms=20, color="g", lw=2, transform=crs, label='matchup point')
    #eras
    ax.quiver(np.array([_dsm["drifter_lon"],_dsm["alti_lon"]]), np.array([_dsm["drifter_lat"],_dsm["alti_lat"]]), np.array([_dsm['es_drifter_matchup_taue'],_dsm['es_alti_matchup_taue']]), np.array([_dsm['es_drifter_matchup_taun'],_dsm['es_alti_matchup_taun']]),
             color='mediumvioletred',transform=crs, lw=3, label='Wind stress',scale=1, width=0.01)

    
    # altimetry
    ax.scatter(_ds["alti_lon"], _ds["alti_lat"], s=np.arange(1, len(_ds["alti_lon"].values)+1)*3, color="orange", transform=crs, label= "satelitte's track")
    ax.plot(_ds["alti_lon"][0], _ds["alti_lat"][0], ".", ms=5, color="orange", transform=crs, label='start')
    ax.plot(_ds["alti_lon"][-1], _ds["alti_lat"][-1], ".", ms=20, color="orange", transform=crs,label='end')
    ax.plot(_dsm["alti_lon"], _dsm["alti_lat"], 
            "*", ms=20, color="darkorange", lw=2, transform=crs, label='matchup point')
    
    gl = ax.gridlines(draw_labels=True, dms=False, 
                 x_inline=False, y_inline=False, 
                )
    gl.right_labels=False
    gl.top_labels=False

    ax.set_title(label + f" obs °{nobs} \n dl={dl/1e3:.1f}km , dt={dt/60:.1f}min , matchup time = "+td.strftime("%Y/%m/%d %H:%M"))
    fig.legend(loc="lower center",  ncol=2, fontsize=10)
    fig.colorbar(C, label ='Sla from Aviso')
    fig.tight_layout(rect=[0,0.2,1,1])
```

```python
plot_coloc_matchup(l, 509)
```
