```python
def plot_closure_bar(ds, id_, ax):
    N=np.arange(0.5,3, 0.5) 
    ax.set_xticks(N, ('sum of all', r'$d_tu$', r'$-fv$', r"$g\partial_x\eta$", r'$\frac{1}{\rho}\partial_z\tau_y$'), fontsize=12)

    #TOTAL
    ax.bar(0.5, ds['rms_sum_'+id_]**2, color ='k',width = 0.5, zorder=3)

    # ACC X
    ax.bar(1, ds['rms_exc_acc_' + id_]**2,
           color = 'lightsteelblue', width = 0.4, zorder=3, align = 'center')
    ax.bar(1,ds['rms_drifter_acc_x']**2, bottom = ds['rms_exc_acc_' + id_]**2,
           color = 'red', width = 0.4, zorder=3, align = 'center')

    # CORIOLIS
    ax.bar(1.5, ds['rms_exc_coriolis_' + id_]**2,
           color = 'lightsteelblue', width = 0.4, zorder=3, align = 'center')
    ax.bar(1.5, ds['rms_drifter_coriolis_x']**2, bottom = ds['rms_exc_coriolis_' + id_]**2,
           color = 'green', width = 0.4, zorder=3, align = 'center')

    # G GRADIENT SLA
    ax.bar(2, ds['rms_exc_ggrad_' + id_]**2,
           color = 'lightsteelblue', width = 0.4, zorder=3, align = 'center')
    ax.bar(2, ds['rms_alti_g_grad_x']**2, bottom = ds['rms_exc_ggrad_' + id_]**2,
           color = 'c', width = 0.4, zorder=3, align = 'center')

    # WIND
    ax.bar(2.5, ds['rms_exc_wind_' + id_]**2,
           color = 'lightsteelblue', width = 0.4, zorder=3, align = 'center')
    ax.bar(2.5, ds['rms_es_cstrio_z15_drifter_wd_x']**2, bottom = ds['rms_exc_wind_' + id_]**2,
           color = 'mediumvioletred', width = 0.4, zorder=3, align = 'center')

    ax.set_ylabel(r'Mean square $\langle ^2 \rangle$ [$m^2/s^4$]')
    ax.set_title(id_)
    #ax.set_ylim((0,8e-5))
    ax.grid(axis='y', zorder=0)

```

```python
def plot_comp_closure_bar(ds, ID, fig_title=1):
    nrows = ceil(len(ID)/2)
    ncols = 2
    # Define the figure and each axis for the 3 rows and 3 columns
    fig, axs = plt.subplots(nrows=nrows,ncols=ncols,
                            sharey=True,
                            figsize=(18,nrows*4))

    # axs is a 2 dimensional array of `GeoAxes`.  We will flatten it into a 1-D array
    axs=axs.flatten()
    

    #Loop over all of the variables
    for i in range(len(ID)):
        plot_closure_bar(ds, ID[i],axs[i])
    if isinstance(fig_title, str):
        fig.suptitle(fig_title)

    # Delete the unwanted axes
    for i in np.arange(len(ID),nrows*2):
        fig.delaxes(axs[i])
        
    colors=['k', 'lightsteelblue']#, 'red', 'green', 'c','mediumvioletred']
    labels_ = [r'$S$',
               r'$S_{-x}$']#,
              #r'$x=d_tu$ from drifters',
              #r'$x=-fv$ from drifters',
              #r"x=$g\partial_x\eta$ from altimeters",
              #r'$x=\frac{1}{\rho}\partial_z\tau_y$ from wind duct']                                                                                                              

    handles = [plt.Rectangle((0,0),1,1, color=colors[i]) for i in range(len(labels_))]
    fig.legend(handles, labels_, ncol=1, fontsize=10,loc='center left', bbox_to_anchor=(1, 0.5))#bbox_transform=fig.transFigure,
    fig.tight_layout(rect=[0,0,1,1])#left, bottom, right, top (default is 0,0,1,1)
```
