## Load and compute 

```python
for l in labels[1:5]:
    ds_data = xr.open_zarr(cstes.zarr_dir+'/'+l+'.zarr').chunk({'obs':500, 'alti_time':-1, 'site_obs':-1}).persist()
    wait(ds_data)
    ds_corr = corr.compute_corr_sla(ds_data).persist()
    wait(ds_corr)
    zarr = os.path.join(zarr_dir, "slacorrection_"+l+".zarr")
    ds_corr.to_zarr(zarr, mode="w")  
    print(f"correction {l} storred in {zarr}")
```
