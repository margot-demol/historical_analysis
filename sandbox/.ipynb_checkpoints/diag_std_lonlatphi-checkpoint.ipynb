## Sum of all terms

```python
ID =['co_denoised_es_cstrio_z0_drifter_x','co_adt_es_cstrio_z0_drifter_x','co_adt_oceantide_es_cstrio_z0_drifter_x']

fig, ax = plt.subplots(3,len(ID), figsize=(20,16 ))
ax.add_feature(cfeature.LAND, zorder=2)
for i in range(len(ID)):
    ds['mean_sum_'+ID[i]].assign_attrs({'long_name':r'$ \langle S \rangle $'}).plot(x='lon_bin', y='lat_bin',cmap='RdBu', ax=ax[0,i])#, vmin=-2e-5, vmax=1.5e-5) 
    ds['std_sum_'+ID[i]].assign_attrs({'long_name':r'$ \sigma(S)$'}).plot(x='lon_bin', y='lat_bin',  cmap='viridis', ax=ax[1,i], vmax=5e-5)
    ds['rms_sum_'+ID[i]].assign_attrs({'long_name':r'$ rms(S)$'}).plot(x='lon_bin', y='lat_bin',  cmap='viridis', ax=ax[2,i], vmax=5e-5)
    ax[0,i].set_title(ID[i])
fig.suptitle('Sum of all terms S')
fig.tight_layout(rect=[0,0,1,0.98])#left, bottom, right, top (default is 0,0,1,1)
```
