```python
def gaussian_pdf(x, mu, sig): 
    return 1/(sig*np.sqrt(2*np.pi))*np.exp(-((x-mu)/sig)**2/2)

def exp_pdf_sig(X,mu,sig):
    lamb=1/sig
    l=[]
    for x in X :
        if x>=0:
            l.append(lamb*np.exp(-lamb*(x-mu))/2)
        else:
            l.append(lamb*np.exp(lamb*(x-mu))/2)

    rescale = np.sum((X[1]-X[0])*np.array(l))
    return np.array(l/rescale)

def gauss_exp(x, mug, sigg, mue, sige):
    return gaussian_pdf(x, mug, sigg) + exp_pdf_sig(x,mue,sige)
```

```python
def fit_gauss(darray, ax,bin_name='v_bin', units='m/s', **kwargs):
    #dh = histogram(darray.rename('vbin'), bins=bins, density=True)
    a = darray.curvefit(coords=darray[bin_name],func= gaussian_pdf)
    fit_a = gaussian_pdf(darray[bin_name].values,  mu=a.sel(param='mu').curvefit_coefficients.values, sig=a.sel(param='sig').curvefit_coefficients.values)
    mu=np.format_float_scientific(a.sel(param='mu').curvefit_coefficients.values, precision = 2, exp_digits=1)
    std_mu=np.format_float_scientific(np.sqrt(a.sel(cov_i='mu',cov_j='mu').curvefit_covariance).values, precision = 1, exp_digits=1)
    sig=np.format_float_scientific(a.sel(param='sig').curvefit_coefficients.values, precision = 2, exp_digits=1)
    std_sig = np.format_float_scientific(np.sqrt(a.sel(cov_i='sig',cov_j='sig').curvefit_covariance).values, precision = 1, exp_digits=1)
    ax.plot(darray[bin_name].values, fit_a, **kwargs)
    #return fr'$\mu$={mu} $\pm$ {std_mu}'+' \n'+fr'$\sigma$= {sig} $\pm$ {std_sig}'
    return r'$\sigma_{gauss}$='+fr'{sig} $\pm$ {std_sig} '+units
```

```python
def fit_exp(darray, ax, bin_name='v_bin', units='m/s', **kwargs):
    #dh = histogram(darray.rename('vbin'), bins=bins, density=True)
    a = darray.curvefit(coords=darray[bin_name],func= exp_pdf_sig)
    fit_a = exp_pdf_sig(darray[bin_name].values,  mu=a.sel(param='mu').curvefit_coefficients.values, sig=a.sel(param='sig').curvefit_coefficients.values)
    mu=np.format_float_scientific(a.sel(param='mu').curvefit_coefficients.values, precision = 2, exp_digits=1)
    std_mu=np.format_float_scientific(np.sqrt(a.sel(cov_i='mu',cov_j='mu').curvefit_covariance).values, precision = 0, exp_digits=1)
    sig=np.format_float_scientific(a.sel(param='sig').curvefit_coefficients.values, precision = 2, exp_digits=1)
    std_sig = np.format_float_scientific(np.sqrt(a.sel(cov_i='sig',cov_j='sig').curvefit_covariance).values, precision = 0, exp_digits=1)
    ax.plot(darray[bin_name].values, fit_a, **kwargs)
    #return fr'$\mu$={mu} $\pm$ {std_mu}'+' \n'+fr'$\sigma$= {sig} $\pm$ {std_sig}'
    return r'$\sigma_{exp}$='+fr'{sig} $\pm$ {std_sig} '+units
```

```python
def fit_gauss_exp(darray, ax, bin_name='v_bin', **kwargs):

    a = darray.curvefit(coords=darray[bin_name], func= gauss_exp)
    fit_a = gauss_exp(darray[bin_name].values,  mug=a.sel(param='mug').curvefit_coefficients.values, sigg=a.sel(param='sigg').curvefit_coefficients.values,
                                      mue=a.sel(param='mue').curvefit_coefficients.values, sige=a.sel(param='sige').curvefit_coefficients.values)
    mug=np.format_float_scientific(a.sel(param='mug').curvefit_coefficients.values, precision = 2, exp_digits=1)
    std_mug=np.format_float_scientific(np.sqrt(a.sel(cov_i='mug',cov_j='mug').curvefit_covariance).values, precision = 2, exp_digits=1)
    sigg=np.format_float_scientific(a.sel(param='sigg').curvefit_coefficients.values, precision = 2, exp_digits=1)
    std_sigg = np.format_float_scientific(np.sqrt(a.sel(cov_i='sigg',cov_j='sigg').curvefit_covariance).values, precision = 0, exp_digits=1)
    mue=np.format_float_scientific(a.sel(param='mue').curvefit_coefficients.values, precision = 2, exp_digits=1)
    std_mue=np.format_float_scientific(np.sqrt(a.sel(cov_i='mue',cov_j='mue').curvefit_covariance).values, precision = 2, exp_digits=1)
    sige=np.format_float_scientific(a.sel(param='sige').curvefit_coefficients.values, precision = 2, exp_digits=1)
    std_sige = np.format_float_scientific(np.sqrt(a.sel(cov_i='sige',cov_j='sige').curvefit_covariance).values, precision = 0, exp_digits=1)
    ax.plot(darray[bin_name].values, fit_a, **kwargs)
    #return fr'$\mu$={mu} $\pm$ {std_mu}'+' \n'+fr'$\sigma$= {sig} $\pm$ {std_sig}'
    return r'$\sigma_{exp}$='+fr'{sige} $\pm$ {std_sige} [m/s]'+'\n'+r'$\sigma_{gauss}$='+fr'{sigg} $\pm$ {std_sigg} [m/s]'
```
