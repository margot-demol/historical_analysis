## Compute gradient on matchup positions (drifter + alti)

```python
def get_matchup_aviso_one_obs(ds_obs):
    """ load aviso for one collocation for matchup time and give aviso g_grad_sla for altimetry position and drifter position at this matchup time"""
    
    # load data
    aviso = sat.load_aviso(ds_obs.time.values, suffix="aviso_")#aviso for matchup time
    
    # interpolate on local grid
    sla = aviso.sla.interp(aviso_lon=ds_obs.box_lon,
                           aviso_lat=ds_obs.box_lat,
                          )

    
    # convert to dataset and massage in order to concatenate properly in get_aviso
    _ds = sla.to_dataset().rename(sla="aviso_sla").isel(aviso_time=0)

    #compute g_grad
    g=9.81
    _ds['aviso_g_grad_x'] = g * _ds.aviso_sla.differentiate('box_x')
    _ds['aviso_g_grad_y'] = g * _ds.aviso_sla.differentiate('box_y')
      
    # matchup site drifter/sat
    assert ds_obs.site_obs.size == 1489, ds_obs
    ds_matchup = ds_obs.isel(site_obs=ds_obs.__site_matchup_indice.compute(),
                             alti_time=10, alti_time_mid=10).compute()
    
    
    _ds['aviso_drifter_matchup_sla'] = _ds.aviso_sla.interp(box_x = ds_matchup.drifter_x.values,
                                                         box_y = ds_matchup.drifter_y.values)
    _ds['aviso_drifter_matchup_g_grad_x'] = _ds.aviso_g_grad_x.interp(box_x = ds_matchup.drifter_x.values,
                                                         box_y = ds_matchup.drifter_y.values)
    _ds['aviso_drifter_matchup_g_grad_y'] = _ds.aviso_g_grad_y.interp(box_x = ds_matchup.drifter_x.values,
                                                         box_y = ds_matchup.drifter_y.values)
    _ds['aviso_alti_matchup_sla'] = _ds.aviso_sla.interp(box_x = ds_matchup.alti_x.values,
                                                         box_y =ds_matchup.alti_y.values)
    _ds['aviso_alti_matchup_g_grad_x'] = _ds.aviso_g_grad_x.interp(box_x = ds_matchup.alti_x.values,
                                                         box_y =ds_matchup.alti_y.values)
    _ds['aviso_alti_matchup_g_grad_y'] = _ds.aviso_g_grad_y.interp(box_x = ds_matchup.alti_x.values,
                                                         box_y =ds_matchup.alti_y.values)

    _ds = (_ds
          .drop(["aviso_time", "time", "lon", "lat", "aviso_lon", "aviso_lat"]) # forget about actual dates in time coordinate
         )#.rename(sla="aviso_sla")
   # )
    return _ds

def get_matchup_aviso(ds):
    """ Load aviso data for multiple collocations and concatenate """
    return xr.concat([get_matchup_aviso_one_obs(ds.sel(obs=o)) for o in ds.obs], 'obs')


def compute_matchup_aviso_sla(ds):
    """
    https://xarray.pydata.org/en/stable/user-guide/dask.html
    Parameters
    ----------
    ds: xr.Dataset
        Input collocation dataset
    dt: tuple
        Time offsets compared to the collocation, e.g. if collocation
        is at time t and dt=(-5,5), aviso data will be interpolated on 
        the interval (t-5 days, t+5 days)
        Default is (-5,5)

    Return
    ------
    ds: xr.Dataset
        Dataset containing relevant aviso data

    """

    # build template
    template = get_matchup_aviso(ds.isel(obs=slice(0,2)))

    # broad cast along obs dimension
    template, _ = xr.broadcast(template.isel(obs=0).chunk(), 
                               ds,
                               exclude=["alti_time", "alti_time_mid", "site_obs", "box_x", "box_y",],
                          )
    # massage further and unify chunks
    template = (xr.merge([template, ds.drop(["lon", "lat", "time"])])
            .unify_chunks()
            [["aviso_sla", "aviso_g_grad_x", "aviso_g_grad_y",
              "aviso_drifter_matchup_sla", "aviso_drifter_matchup_g_grad_x","aviso_drifter_matchup_g_grad_y",
              "aviso_alti_matchup_sla", "aviso_alti_matchup_g_grad_x","aviso_alti_matchup_g_grad_y",
             ]]
            )
    
    # dimension order is not consistent with that of exiting from map_blocks for some reason
    template["aviso_sla"] = template["aviso_sla"].transpose("obs", "box_y", "box_x")
    template["aviso_g_grad_x"] = template["aviso_g_grad_x"].transpose("obs", "box_y", "box_x")
    template["aviso_g_grad_y"] = template["aviso_g_grad_y"].transpose("obs", "box_y", "box_x")
    
    #chunk alti_time et site_obs in only one chunk both
    ds=ds.chunk({'alti_time':ds_data.alti_time.size, 'site_obs':ds_data.site_obs.size})
    
    # actually perform the calculation
    ds_sla = xr.map_blocks(get_matchup_aviso, ds, template=template)
    
    return ds_sla
```
